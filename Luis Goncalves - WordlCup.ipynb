```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np

cup = pd.read_csv("WorldCups.csv")
cup
```

```python
cup.dtypes
```

```python
sns.set()

cup_filtre=cup[cup["Year"]>1980]
cup_filtre=cup[cup["QualifiedTeams"]<20]
sns.relplot(x='Year', y='QualifiedTeams', data=cup_filtre)

cup_filtre

#Dans ce graphique ou peut voir combien d'équipes se sont qualifiés pour la coupe du monde depuis le début de l'édition World Cup jusqu'à 1980.
```

```python
sns.set()

cup_filtre=cup[cup["Year"]<1980]
cup_filtre=cup[cup["QualifiedTeams"]>20]
sns.relplot(x='Year', y='QualifiedTeams', data=cup_filtre)

cup_filtre

#Dans ce graphique on constate qu'entre de 1980 à 1994 il y a eu 24 équipes qui se sont qualifiés pour les 4 éditions de coupe du monde dans cette période.
#Ainsi qu'apartir de 1995 jusqu'à 2015 il y a eu 32 équipes qui se sont qualifiés pour les 5 éditions de coupe du monde dans cette période
```

```python
sns.set()

cup_filtre=cup[cup["GoalsScored"]<30]
cup_filtre=cup[cup["MatchesPlayed"]>20]
sns.relplot(x='GoalsScored', y='MatchesPlayed', data=cup_filtre)

cup_filtre

#Dans ce graphique on peut voir le nombre de buts qu'il y a eu par rapport au nombre de matchs joués, avec un filtre de minimum 30 buts pour plus de 20 matchs.
```
